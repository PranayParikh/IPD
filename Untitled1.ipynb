```python
from fastapi import FastAPI
from pydantic import BaseModel
from typing import List
import uvicorn

app = FastAPI(title="Ayusphere Chatbot API (Colab)")

class ChatRequest(BaseModel):
    message: str

class ChatResponse(BaseModel):
    reply: str


def classify_risk(text: str) -> str:
    t = text.lower()

    emergency_flags = [
        "chest pain",
        "can't breathe",
        "cannot breathe",
        "difficulty breathing",
        "short of breath",
        "shortness of breath",
        "severe pain",
        "very bad pain",
        "crushing pain",
        "fainted",
        "lost consciousness",
        "vomiting blood",
        "passing blood",
        "heavy bleeding",
        "stroke",
        "face drooping",
        "slurred speech",
        "suicidal",
        "want to die",
        "kill myself",
        "pregnant and bleeding",
        "pregnancy and bleeding",
    ]

    if any(flag in t for flag in emergency_flags):
        return "emergency"
    return "normal"


def generate_self_care(text: str) -> dict:
    t = text.lower()

    # HEADACHE
    if "headache" in t or "head pain" in t:
        return {
            "summary": "It sounds like you’re having a mild headache.",
            "steps": [
                "Drink some water and see if dehydration could be a cause.",
                "Rest your eyes and avoid long screen time for a while.",
                "Sit or lie down in a quiet, dim room if possible.",
                "Notice if anything triggers it (stress, screens, loud noise, skipped meals).",
                "If the headache becomes very severe, is sudden and worst-ever, or you have confusion, weakness, or trouble speaking, please seek urgent medical care.",
            ],
        }

    # FEVER / COLD
    if "fever" in t or "cold" in t or "cough" in t:
        return {
            "summary": "It sounds like you may have mild fever or cold-like symptoms.",
            "steps": [
                "Stay well hydrated with water, ORS, or warm fluids like soup or herbal tea.",
                "Get plenty of rest and avoid overexerting yourself.",
                "Use simple home comfort measures like steam inhalation and warm salt water gargles for throat (be careful with hot water).",
                "Monitor your temperature and how you feel over the next 24–48 hours.",
                "See a doctor if fever is very high, lasts more than 2–3 days, you have chest pain, breathing difficulty, or feel very unwell.",
            ],
        }

    # STOMACH ISSUES
    if (
        "stomach" in t
        or "abdomen" in t
        or "abdominal" in t
        or "vomit" in t
        or "diarrhea" in t
        or "diarrhoea" in t
        or "loose motion" in t
    ):
        return {
            "summary": "It sounds like you’re having some stomach or digestion-related discomfort.",
            "steps": [
                "Sip fluids regularly to avoid dehydration (water, ORS, clear soups).",
                "Eat light, simple foods if you can tolerate them (like khichdi, banana, toast).",
                "Avoid very oily, spicy, or heavy foods for a while.",
                "If you have severe, sharp pain, persistent vomiting, blood in stool or vomit, or fever with strong pain, please see a doctor or emergency care urgently.",
            ],
        }

    # STRESS / ANXIETY / LOW MOOD
    if (
        "anxiety" in t
        or "anxious" in t
        or "stressed" in t
        or "stress" in t
        or "sad" in t
        or "low mood" in t
    ):
        return {
            "summary": "It sounds like you’re feeling emotionally stressed or low.",
            "steps": [
                "Pause and take 5 slow deep breaths, inhaling through the nose and exhaling through the mouth.",
                "Try a grounding exercise: notice 5 things you can see, 4 things you can touch, 3 things you can hear, 2 things you can smell, 1 thing you can taste.",
                "If you can, talk to a trusted friend or family member about how you’re feeling.",
                "Consider journaling your thoughts for a few minutes to get them out of your head.",
                "If these feelings are very strong, persistent, or include thoughts of self-harm, please seek help from a mental health professional or emergency services.",
            ],
        }

    # DEFAULT
    return {
        "summary": "I understand you’re not feeling well.",
        "steps": [
            "Try to rest and notice which symptoms are bothering you the most.",
            "Stay hydrated with water or other fluids unless a doctor has told you to restrict fluids.",
            "Avoid very heavy, oily, or junk food while you’re feeling unwell.",
            "If symptoms get worse, last many days, or you’re worried, please see a doctor for a proper check-up.",
        ],
    }


@app.post("/chat", response_model=ChatResponse)
def chat(req: ChatRequest):
    user_message = (req.message or "").strip()

    if not user_message:
        return ChatResponse(
            reply="Please tell me a bit about how you’re feeling so I can guide you."
        )

    risk = classify_risk(user_message)

    if risk == "emergency":
        reply_lines: List[str] = [
            "Your symptoms could be serious.",
            "Please seek immediate medical help or go to the nearest emergency department.",
            "Do not rely on chat or home remedies in this situation.",
            "",
            "I am a virtual assistant, not a doctor.",
        ]
        return ChatResponse(reply="\n".join(reply_lines))

    data = generate_self_care(user_message)
    summary = data["summary"]
    steps = data["steps"]

    reply = (
        summary
        + "\n\nHere are a few gentle, general steps you can try:\n"
        + "\n".join(f"{i+1}. {step}" for i, step in enumerate(steps))
        + "\n\nI am a virtual assistant and not a doctor. This is general information, not a diagnosis or medical advice. Please consult a healthcare professional for personalised care."
    )

    return ChatResponse(reply=reply)
```

```python
from fastapi import FastAPI
from pydantic import BaseModel
from typing import List
import uvicorn

app = FastAPI(title="Ayusphere Chatbot API (Colab)")

class ChatRequest(BaseModel):
    message: str

class ChatResponse(BaseModel):
    reply: str


# --- Simple risk classifier ---
def classify_risk(text: str) -> str:
    t = text.lower()

    emergency_flags = [
        "chest pain",
        "can't breathe",
        "cannot breathe",
        "difficulty breathing",
        "short of breath",
        "shortness of breath",
        "severe pain",
        "very bad pain",
        "crushing pain",
        "fainted",
        "lost consciousness",
        "vomiting blood",
        "passing blood",
        "heavy bleeding",
        "stroke",
        "face drooping",
        "slurred speech",
        "suicidal",
        "want to die",
        "kill myself",
        "pregnant and bleeding",
        "pregnancy and bleeding",
    ]

    if any(flag in t for flag in emergency_flags):
        return "emergency"

    return "normal"


# --- Symptom-based helper ---
def generate_self_care(text: str) -> dict:
    t = text.lower()

    # ⚠️ CHEST PAIN:
    # Chest pain is handled as an emergency in classify_risk()
    # on purpose, so we do NOT add a mild self-care block here.
    # If "chest pain" appears, classify_risk() will send the
    # user to urgent care instead of giving home remedies.

    # HEADACHE
    if "headache" in t or "head pain" in t:
        return {
            "summary": "It sounds like you’re having a mild headache.",
            "steps": [
                "Drink some water and see if dehydration could be a cause.",
                "Rest your eyes and avoid long screen time for a while.",
                "Sit or lie down in a quiet, dim room if possible.",
                "Notice if anything triggers it (stress, screens, loud noise, skipped meals).",
                "If the headache becomes very severe, is sudden and worst-ever, or you have confusion, weakness, or trouble speaking, please seek urgent medical care.",
            ],
        }

    # FEVER / COLD
    if "fever" in t or "cold" in t or "cough" in t:
        return {
            "summary": "It sounds like you may have mild fever or cold-like symptoms.",
            "steps": [
                "Stay well hydrated with water, ORS, or warm fluids like soup or herbal tea.",
                "Get plenty of rest and avoid overexerting yourself.",
                "Use simple home comfort measures like steam inhalation and warm salt water gargles for throat (be careful with hot water).",
                "Monitor your temperature and how you feel over the next 24–48 hours.",
                "See a doctor if fever is very high, lasts more than 2–3 days, you have chest pain, breathing difficulty, or feel very unwell.",
            ],
        }

    # STOMACH ISSUES
    if (
        "stomach" in t
        or "abdomen" in t
        or "abdominal" in t
        or "vomit" in t
        or "vomiting" in t
        or "diarrhea" in t
        or "diarrhoea" in t
        or "loose motion" in t
    ):
        return {
            "summary": "It sounds like you’re having some stomach or digestion-related discomfort.",
            "steps": [
                "Sip fluids regularly to avoid dehydration (water, ORS, clear soups).",
                "Eat light, simple foods if you can tolerate them (like khichdi, banana, toast).",
                "Avoid very oily, spicy, or heavy foods for a while.",
                "If you have severe, sharp pain, persistent vomiting, blood in stool or vomit, or fever with strong pain, please see a doctor or emergency care urgently.",
            ],
        }

    # NAUSEA (can appear without vomiting/diarrhea)
    if "nausea" in t or "nauseous" in t or "feel like vomiting" in t or "feeling like vomiting" in t:
        return {
            "summary": "It sounds like you’re feeling nauseous.",
            "steps": [
                "Take small sips of water or ORS frequently instead of large gulps.",
                "Avoid very strong smells, oily or spicy foods for a while.",
                "Try bland foods like dry toast, crackers, or plain khichdi if you feel like eating.",
                "Sit upright or keep your head slightly elevated instead of lying completely flat.",
                "See a doctor if you cannot keep any fluids down, the nausea is severe or lasts many hours, you have chest pain, severe stomach pain, or if you might be pregnant.",
            ],
        }

    # BACK PAIN
    if "back pain" in t or "backache" in t or "lower back" in t or "upper back" in t:
        return {
            "summary": "It sounds like you’re having some back pain or backache.",
            "steps": [
                "Rest for short periods, but avoid lying completely still for the whole day — gentle movement often helps.",
                "Try to maintain a comfortable posture; avoid slouching or awkward positions for long.",
                "If it feels comfortable for you, you can try a warm compress or heating pad on the painful area (not too hot, and not directly on bare skin).",
                "Avoid lifting heavy objects or sudden twisting movements until the pain settles.",
                "See a doctor urgently if the pain is very severe, travels down your legs, you have weakness or numbness, trouble controlling urine or stool, or you have fever along with back pain.",
            ],
        }

    # STRESS / ANXIETY / LOW MOOD
    if (
        "anxiety" in t
        or "anxious" in t
        or "stressed" in t
        or "stress" in t
        or "sad" in t
        or "low mood" in t
    ):
        return {
            "summary": "It sounds like you’re feeling emotionally stressed or low.",
            "steps": [
                "Pause and take 5 slow deep breaths, inhaling through the nose and exhaling through the mouth.",
                "Try a grounding exercise: notice 5 things you can see, 4 things you can touch, 3 things you can hear, 2 things you can smell, 1 thing you can taste.",
                "If you can, talk to a trusted friend or family member about how you’re feeling.",
                "Consider journaling your thoughts for a few minutes to get them out of your head.",
                "If these feelings are very strong, persistent, or include thoughts of self-harm, please seek help from a mental health professional or emergency services.",
            ],
        }

    # DEFAULT
    return {
        "summary": "I understand you’re not feeling well.",
        "steps": [
            "Try to rest and notice which symptoms are bothering you the most.",
            "Stay hydrated with water or other fluids unless a doctor has told you to restrict fluids.",
            "Avoid very heavy, oily, or junk food while you’re feeling unwell.",
            "If symptoms get worse, last many days, or you’re worried, please see a doctor for a proper check-up.",
        ],
    }



@app.post("/chat", response_model=ChatResponse)
def chat(req: ChatRequest):
    user_message = (req.message or "").strip()

    if not user_message:
        return ChatResponse(reply="Please describe how you’re feeling.")

    risk = classify_risk(user_message)

    if risk == "emergency":
        return ChatResponse(
            reply="Your symptoms could be serious.\nPlease seek emergency medical help.\n\nI am not a doctor."
        )

    data = generate_self_care(user_message)
    summary = data["summary"]
    steps = data["steps"]

    reply = (
        summary
        + "\n\nHere are some safe things to try:\n"
        + "\n".join(f"{i+1}. {step}" for i, step in enumerate(steps))
        + "\n\n(I am a virtual assistant, not a doctor.)"
    )

    return ChatResponse(reply=reply)
```

```python
from threading import Thread

def run_server():
    uvicorn.run(app, host="0.0.0.0", port=8000)

thread = Thread(target=run_server, daemon=True)
thread.start()

print("✅ Ayusphere chatbot server started on port 8000")
```
